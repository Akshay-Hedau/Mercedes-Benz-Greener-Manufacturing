# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
cat_col = list(df.select_dtypes(include="object").columns)
cat_col
```

```python
for col in cat_col:
    df[col]=le.fit_transform(df[col])
```

```python
# All categorical object colu mns has been converted to numeric columns.
```

```python
cols_to_drop = []

for cols in df.columns:
    if df[cols].var() == 0:
        cols_to_drop.append(cols)
```

```python
cols_to_drop
```

```python
# Columns with no variance
df[cols_to_drop].nunique()
```

```python
df = df.drop(cols_to_drop,axis=1)
```

# Using Ordinary Least Square Method to Find Out Features Significant to Target Variable:

```python
X = df.drop('y', axis = 1)
y = df.y
```

```python
X.shape
```

```python
y.shape
```

```python
from statsmodels.api import OLS
```

```python
ols_model = OLS(y,X)
```

```python
result = ols_model.fit()
```

```python
result.summary()
```

```python
# Lot of Columns have P-Value > 0.05.

# These columns are not required to Predict Target.

# We can try Building the Model Without these Columns.
```

```python
result.pvalues
```

```python
result.pvalues[result.pvalues<0.05]
```

```python
result.pvalues[result.pvalues<0.05].index
```

```python
significant_cols = list(result.pvalues[result.pvalues<0.05].index)
```

```python
significant_cols
```

```python
len(significant_cols)
```

```python
# We will use only these 70 columns for our model building.
```

```python
X_new = X[significant_cols].copy()
```

```python
X_new.shape
```

```python
X_new.head()
```

# Scalling:
While Mostly all Categorical Variables have Values of 0 and 1, Variables that we did Label Encoder on have Multiple Values. So, it is Possible that Machine Learning Models will give More Importance to them as They have Values Greater Than 1. So, we will Scale the dataset.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
temp = sc.fit_transform(X_new)
scaled_x = pd.DataFrame(temp, index=X_new.index, columns= X_new.columns)
scaled_x.head()
```

# Dimensionality Reduction Using PCA:

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.99)
```

```python
pca_result = pca.fit_transform(scaled_x)
```

```python
pca_result
```

```python
pca_result.shape
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```
